### 문제1) 아래 문장에서 각각의 단어가 몇 번이나 나오는지를 계산하는 프로그램을 작성하기

눈물을... 흘.린.ㄷㅏ...(한♤숨)

```python
#한줄씩 파일 읽는건 못해서 한줄로 수정했음 이해 바람
str='크큭... ㄴr는 to day도 ㅇㅏ름ㄷㅏ운 달을 바ㄹr보며... 눈물을... 흘.린.ㄷㅏ...(한♤숨) ㅈㅓ 달은 ㄴㅏ의 고.도.쿠☆한 ㅎㅏ트를 진정ㅅlㅋㅕ주는 ㄴㅏ의 angel♡ 그ㄷㅐ의 뷰ㅌㅣ풀한 트윙클☆을 ㅂㅏㄹr봄에... 와타시는 투ㄷㅔE도 아픈 ♡를 달랜ㄷr...'

word_list=str.split( )
word={}
word_count={}
for each in word_list:
        if each in word_count: 
                word_count[each]+=1
        else:
                word_count[each]=1
 
for k,v, in word_count.items():
    print(k,v)
```

```python
data='''크큭... ㄴr는 to day도 ㅇㅏ름ㄷㅏ운 달을 바ㄹr보며...

눈물을... 흘.린.ㄷㅏ...(한♤숨)

ㅈㅓ 달은 ㄴㅏ의 고.도.쿠☆한 ㅎㅏ트를 진정ㅅlㅋㅕ주는 ㄴㅏ의 angel♡

그ㄷㅐ의 뷰ㅌㅣ풀한 트윙클☆을 ㅂㅏㄹr봄에...

와타시는 투ㄷㅔE도 아픈 ♡를 달랜ㄷr...'''

from konlpy.tag import Okt
from konlpy.utils import pprint
kkma = Okt()
lines=data.split('\n')
freq = {}
for line in lines :
    tokens=kkma.nouns(line)
    pprint(kkma.pos(line))
    for token in tokens:
        if token in freq.keys():
            freq[token] = freq[token]+1
        else :
            freq[token] = 1
freq
```

### 문제3)
아래 두개의 list에서 
1을 입력시 교집합을 리턴
2을 입력시 합집합을 리턴
3을 입력시 여집합을 리턴
하는 합수를 람다식으로 만드세여<br>
A=\{1,2,7,15,23}<br>
B=\{2,15,16,27}

```python
A={1,2,7,15,23}
B={2,15,16,27}

def fun(a):
 if a==1:
    return A & B
 elif a==2:
    return A | B
 elif a==3:
    return A - B
```

```python
fun(3)
```
